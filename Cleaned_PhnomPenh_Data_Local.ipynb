```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Read Excel from HDFS") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://localhost:9000") \
    .config("spark.jars.packages", "com.crealytics:spark-excel_2.12:0.13.4") \
    .getOrCreate()

# Path to the Excel file on HDFS
base_path = 'hdfs://localhost:9000/user/hadoop/mydata'
hdfs_path = f'{base_path}/merged_output.xlsx'

# Read the Excel file into a DataFrame
df = spark.read.format("com.crealytics.spark.excel") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("dataAddress", "Sheet1") \
    .load(hdfs_path)
```

```python
# importing the required modules
import glob
# Import the neccessary modules for data manipulation and visual representation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as matplot
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
path = "data/phnom_penh"
# excel files in the path
file_list = glob.glob(path + "/*.xlsx")
print(file_list)
```

```python
# list of excel files we want to merge.
# pd.read_excel(file_path) reads the  
# excel data into pandas dataframe.
excl_list = []
 
for file in file_list:
    excl_list.append(pd.read_excel(file))
 
# concatenate all DataFrames in the list
# into a single DataFrame, returns new 
# DataFrame.
excl_merged = pd.concat(excl_list, ignore_index=True)
 
# exports the dataframe into excel file
# with specified name.
excl_merged.to_excel('MERGED_PHNOM_PENH_DATASET.xlsx', index=False)
```
